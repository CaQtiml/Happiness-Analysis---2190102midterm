## **Spearman's rank correlation**

Spearman's rank correlation is, oversimplifyingly, a mean of measuring how much two variable correlate with each other.

Spearman's rank correlation coefficient can interpret how much two variable are related to each other, with an assumption that the datas are
- Monotonic: If the size of one variable increase/decrease, then the other must also increase/decrease, regardless of whether the function is linear.
- Ordinal: Rankable, regardless of whether the distance between each rank is known.

Spearman's rank correlation coefficient's value can rank between 1 to -1, with 1 means the variable perfectly and positively correlate with each other (one increase, the other increase) and -1 means otherwise (one increase, the other decrease)

## **$ρ_{Spearman} = 1- {\frac{6}{n(n^2-1)}} \sum_{i=1}^{n} d^2_i$**

```python
# list of intersected countries
intersected_country = list(set(df6['Country']).intersection(set(dfco2['Country'])))

for i in df6.Country:
  if i not in intersected_country:
    df6 = df6[df6.Country != i]

for i in dfco2.Country:
  if i not in intersected_country:
    dfco2 = dfco2[dfco2.Country != i]
```

```python
df6_spearman = df6[['Region', 'Country', 'Happiness Score']]
df6_spearman = df6_spearman.sort_values(by='Happiness Score')
df6_spearman['Rank_Happiness'] = [x for x in range(1, df6_spearman['Country'].size+1)]
```

```python
dfco2_spearman = dfco2[['Country', 'CO2Emissions']]
dfco2_spearman = dfco2_spearman.sort_values(by='CO2Emissions')
dfco2_spearman['Rank_CO2'] = [x for x in range(1, dfco2_spearman['Country'].size+1)]
```

```python
merged_df_spearman = pd.merge(df6_spearman, dfco2_spearman, how='outer')
merged_df_spearman['d'] = merged_df_spearman['Rank_Happiness'] - merged_df_spearman['Rank_CO2']
merged_df_spearman['d_square'] = merged_df_spearman['d'] ** 2
```

```python
n = len(merged_df_spearman['Country'])

print(f"Spearman's Correlation Coefficient: {1-(6 * merged_df_spearman['d_square'].sum() / (n * ((n ** 2) - 1)))}")
```

## **Pearson's correlation**

Pearson correlation coefficient is the measure of how much two variable linearly correlate with each other.

Pearson correlation has the assumption that the data is
- Continuous: Continuous range of numerical value
- Related pair: Each observation of the data should include both of variable
- Absence of Outlier: No abnormality that defers from the rest of the data
- Linear

Pearson correlation coefficient also range from 1 to -1, with 1 means that the two variable are perfectly, postively and linearly correlate with each other and -1 means otherwise.

##**$r_{Pearson} = \frac{n\sum_{i=1}^n x_iy_i- \sum_{i=1}^n x_i\sum_{i=1}^n y_i}{\sqrt{(n\sum_{i=1}^n x^2_i - (\sum_{i=1}^n x_i)^2)(n\sum_{i=1}^n y^2_i- (\sum_{i=1}^n y_i)^2)}}$**

Note: In pandas, you can also use corr() function to also find the Pearson correlation coefficient.

```python
df6_pearson = df6[['Region', 'Country', 'Happiness Score']]
dfco2_pearson = dfco2[['Country', 'CO2Emissions']]
```

```python
merged_df_pearson = pd.merge(df6_pearson, dfco2_pearson)

merged_df_pearson.columns = ['Region', "Country", 'x', 'y']
for i in ['China', 'United States', 'Japan', 'Russia', 'India']:
  merged_df_pearson = merged_df_pearson[merged_df_pearson.Country != i]

merged_df_pearson['x'] = merged_df_pearson['x'].astype('float128') # Changing data type from float64 or int64 to float128
merged_df_pearson['y'] = merged_df_pearson['y'].astype('float128')
merged_df_pearson['x2'] = merged_df_pearson['x'] ** 2
merged_df_pearson['y2'] = merged_df_pearson['y'] ** 2
merged_df_pearson['xy'] = merged_df_pearson['x'] * merged_df_pearson['y']
n = len(merged_df_pearson.Country)

tt = (n * merged_df_pearson.xy.sum()) - (merged_df_pearson.x.sum() * merged_df_pearson.y.sum())
b1 = math.sqrt((n * merged_df_pearson.x2.sum()) - (merged_df_pearson.x.sum() ** 2))
b2 = math.sqrt((n * merged_df_pearson.y2.sum()) - (merged_df_pearson.y.sum() ** 2))
```

```python
print(f"Pearson's Correlation Coefficient: {tt / (b1 * b2)}")
```

## **Normalized Happiness scores and $CO_2$ Emissions in countries**

To decrease the redundancy of the data and improve datas' integrity, we normalized the data. Normalization is the process of restructuring the database. Normalized value will be between 0 to 1.

## **$x_{normalized} = \frac{x_i-x_{min}}{x_{max}-x_{min}}$**

As we normalized both the Happiness score and the $CO_2$ Emission of each countries, we can plot them against each other in a graph.

```python
del df6_pearson, df6_spearman, dfco2_pearson, dfco2_spearman, merged_df_pearson, merged_df_spearman
```

```python
def standardization(df_, key_): 
  """ 
    Add new DataSeries of 'NORMALIZED' value to the DataFrame df_ and return it

    :param df_: pandas' DataFrame
    :param key_: DataSeries key
    :return df_: DataFrame with new DataSeries of standardized value
  """

  # df_mean = df_[key_].mean()
  # df_std = df_[key_].std()
  # df_[key_ + "_STDized"] = (df_[key_] - df_mean) / df_std

  df_min = df_[key_].min()
  df_max = df_[key_].max()
  df_[key_ + "_STDized"] = (df_[key_] - df_min) / (df_max-df_min)
  return df_

def plot_seaborn(df_, xkey_, ykey_, xlabel_, ylabel_, title_, llabel_, ltitle_, rotate_=90):
  """
    This is the function to be called that completely plot the graph using seaborn

    :param df_: Pandas' DataFrame
    :param xkey_: String of df_ key to be plot on the x axis
    :param ykey_: List of df_ key to be plot on the y axis
    :param xlabel_: String of x axis label
    :param ylabel_: String of y axis label
    :param title_: String of graph title
    :param llabel_: List of legend label
    :param ltitle_: String of legend label
    :param rotate_: Integer of degree of rotation of X-axis label rotation

    :return: 
  """
  fig,ax = plt.subplots(figsize=(24,11))
  for y in ykey_:
    chart = sbs.lineplot(x=df_[xkey_], y=df_[y], data=df_)

  chart.set_xticklabels(list(df_[xkey_]), rotation=rotate_)
  chart.set_xlabel(xlabel_, fontsize=22)
  chart.set_ylabel(ylabel_, fontsize=22)
  chart.set_title("\n" + title_ + "\n", fontsize=24)
  plt.legend(labels=llabel_, fontsize='large', title='Legend', title_fontsize=20,)
  plt.show()

def set_to_mean(df_, countries_):
  """
    This function take in df_ and set the value to mean

    :param df_: the pandas' DataFrame
    :param countries_: list of string of countries you want to modify
    :return df_: modified Pandas' DataFrame 
  """
  h_mean = df_['Happiness Score'].mean()
  c_mean = df_['CO2Emissions'].mean()
  for i in countries_:
    # df_.at[df_[df_.Country==i].index[0], 'Happiness Score'] = h_mean # This line set the Happiness Score to its mean
    df_.at[df_[df_.Country==i].index[0], 'CO2Emissions'] = c_mean
  
  return df_
```

```python
# Merge the happiness and emission data file
merged_df = pd.merge(df6[['Region', 'Country', 'Happiness Score']], dfco2[['Country', 'CO2Emissions']])

# Standardizing data
merged_df_1 = standardization(merged_df, 'Happiness Score')
merged_df_1 = standardization(merged_df, 'CO2Emissions')

# Plot with seaborn
plot_seaborn(df_=merged_df_1, xkey_='Country', ykey_=['Happiness Score_STDized','CO2Emissions_STDized'],
            xlabel_='Country', ylabel_="Standardized Score",
            title_="Normalized Score VS Country", ltitle_="Legend",
            llabel_=['Happiness Score (Standardized)', 'CO2 Emission (Standardized)'])
```

```python
px.scatter(merged_df_1, x='Happiness Score', y='CO2Emissions', 
           color='Country', trendline='ols', trendline_scope='overall',
           width=1280, height=720, title='CO2 Emission VS Happiness Scores')
```
